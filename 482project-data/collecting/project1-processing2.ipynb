```python
import json
import os
import glob
import numpy as np

def convert_date(my_date):
    split_my_date=my_date.strip().split(" ")
    
    months=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
    day=split_my_date[1]
    year=split_my_date[2]
    month=months.index(split_my_date[0])+1
    whole_date=str(month)+"/"+day+"/"+year
    return whole_date




amazon_path=[]
google_path=[]
facebook_path=[]
apple_path=[]

with open("amazon_dates.txt") as f:
    line=f.readline()
    while(line):
        split_line=line.strip().split("/")
        date=split_line[-1]
        my_dictionary="amazon_test/"+date+"_auto_anns/exp_polarity.txt"
        
        amazon_path.append(my_dictionary)
        line=f.readline()
        
with open("google_dates.txt") as f:
    line=f.readline()
    while(line):
        split_line=line.strip().split("/")
        date=split_line[-1]
        my_dictionary="google_test/"+date+"_auto_anns/exp_polarity.txt"
        
        google_path.append(my_dictionary)
        line=f.readline()
        
with open("facebook_dates.txt") as f:
    line=f.readline()
    while(line):
        split_line=line.strip().split("/")
        date=split_line[-1]
        my_dictionary="facebook_test/"+date+"_auto_anns/exp_polarity.txt"
        
        facebook_path.append(my_dictionary)
        line=f.readline()
        
with open("apple_dates.txt") as f:
    line=f.readline()
    while(line):
        split_line=line.strip().split("/")
        date=split_line[-1]
        my_dictionary="apple_test/"+date+"_auto_anns/exp_polarity.txt"
        
        apple_path.append(my_dictionary)
        line=f.readline()
        
amazon_result=[]
for amazon in amazon_path:
    file_name=amazon.strip().split("/")[1]
    date=convert_date(file_name.strip().split(".")[0])
    
    check=0
    result=date+","
    total=0
    pos=0
    neg=0
    neu=0
    with open(amazon) as f:
        line=f.readline()
        while(line):
            
            line_split=line.strip().split("\t")
            sensitive=line_split[1]
            if (sensitive=="neutral"):
                neu+=1
            elif (sensitive =="positive"):
                pos+=1
            elif(sensitive == "negative"):
                neg+=1
            total+=1
            line=f.readline()
        if(total==0):
            result+="0.3333333333333333,0.3333333333333333,0.3333333333333333"
        else:
            result+=str(pos/total)+","
            result+=str(neg/total)+","
            result+=str(neu/total)
        amazon_result.append(result)

apple_result=[]
for apple in apple_path:
    file_name=apple.strip().split("/")[1]
    date=convert_date(file_name.strip().split(".")[0])
    check=0
    result=date+","
    total=0
    pos=0
    neg=0
    neu=0
    with open(apple) as f:
        line=f.readline()
        while(line):
            
            line_split=line.strip().split("\t")
            sensitive=line_split[1]
            if (sensitive=="neutral"):
                neu+=1
            elif (sensitive =="positive"):
                pos+=1
            elif(sensitive == "negative"):
                neg+=1
            total+=1
            line=f.readline()
        if(total==0):
            result+="0.3333333333333333,0.3333333333333333,0.3333333333333333"
        else:
            result+=str(pos/total)+","
            result+=str(neg/total)+","
            result+=str(neu/total)
        apple_result.append(result)
        
facebook_result=[]
for facebook in facebook_path:
    file_name=facebook.strip().split("/")[1]
    date=convert_date(file_name.strip().split(".")[0])
    check=0
    result=date+","
    total=0
    pos=0
    neg=0
    neu=0
    with open(facebook) as f:
        line=f.readline()
        while(line):
            
            line_split=line.strip().split("\t")
            sensitive=line_split[1]
            if (sensitive=="neutral"):
                neu+=1
            elif (sensitive =="positive"):
                pos+=1
            elif(sensitive == "negative"):
                neg+=1
            total+=1
            line=f.readline()
        if(total==0):
            result+="0.3333333333333333,0.3333333333333333,0.3333333333333333"
        else:
            result+=str(pos/total)+","
            result+=str(neg/total)+","
            result+=str(neu/total)
        facebook_result.append(result)

google_result=[]
for google in google_path:
    file_name=google.strip().split("/")[1]
    date=convert_date(file_name.strip().split(".")[0])
    check=0
    result=date+","
    total=0
    pos=0
    neg=0
    neu=0
    with open(google) as f:
        line=f.readline()
        while(line):
            
            line_split=line.strip().split("\t")
            sensitive=line_split[1]
            if (sensitive=="neutral"):
                neu+=1
            elif (sensitive =="positive"):
                pos+=1
            elif(sensitive == "negative"):
                neg+=1
            total+=1
            line=f.readline()
        if(total==0):
            result+="0.3333333333333333,0.3333333333333333,0.3333333333333333"
        else:
            result+=str(pos/total)+","
            result+=str(neg/total)+","
            result+=str(neu/total)
            google_result.append(result)
        
np.savetxt("amazon_results.txt", amazon_result, delimiter=" ",fmt="%s")
np.savetxt("facebook_results.txt", facebook_result, delimiter=" ",fmt="%s")
np.savetxt("apple_results.txt", apple_result, delimiter=" ",fmt="%s")
np.savetxt("google_results.txt", google_result, delimiter=" ",fmt="%s")
```
